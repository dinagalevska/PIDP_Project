#Training the model using ANN(Artificial Neural Network)

```python
import tensorflow as tf
import keras
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
def create_model():
  model = Sequential()
  model.add(Dense(7, activation='relu'))
  model.add(Dropout(0.25))
  model.add(Dense(16, activation='relu'))
  model.add(Dropout(0.25))
  model.add(Dense(32, activation='relu'))
  model.add(Dropout(0.25))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss=keras.losses.BinaryCrossentropy(), optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
  return model
```

```python
model=create_model()
model.fit(X_train, y_train, epochs=10)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
model=create_model()
deep_history = model.fit(X_train,
                              y_train,
                              epochs=50,
                              batch_size = 8,
                              validation_data = (X_test, y_test),
                              callbacks=[early_stop]
                            )
```

```python
historydf = pd.DataFrame(deep_history.history, index=deep_history.epoch)
plt.figure(figsize=(8, 6))
historydf.plot(ylim=(0, historydf.values.max()))
plt.title('Loss: %.3f' % deep_history.history['loss'][-1])
plt.show()
```

#Principal Component Analysis (PCA)


```python
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
scores = []

for n_components in range(1, 5):
    pca = PCA(n_components=n_components)

    X_train_pca = pca.fit_transform(X_train)

    X_test_pca = pca.transform(X_test)

    model=create_model()
    early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
    deep_history = model.fit(X_train_pca,
                                  y_train,
                                  epochs=50,
                                  batch_size = 8,
                                  validation_data = (X_test_pca, y_test),
                                  callbacks=[early_stop]
                                )

    preds = (model.predict(X_test_pca) > 0.8).astype(int)

    acc = accuracy_score(y_test, preds)
    scores.append(acc)
```

```python
for n_components, score in zip(range(1, 5), scores):
    print(f"Components: {n_components}, Accuracy: {score:.4f}")
```

#Paralleize a NN with joblib-across multiple CPU cores

```python
from joblib import Parallel, delayed
import numpy as np
```

```python
def train_model(X_chunk, y_chunk):
    X_train_chunk, X_val_chunk, y_train_chunk, y_val_chunk = train_test_split(
        X_chunk, y_chunk, test_size=0.2, random_state=42
    )

    model = create_model()
    early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
    history = model.fit(
        X_train_chunk,
        y_train_chunk,
        epochs=50,
        batch_size=8,
        validation_data=(X_val_chunk, y_val_chunk),
        callbacks=[early_stop]
    )
    return history

num_cores = 2
X_train_chunks = np.array_split(X_train, num_cores)
y_train_chunks = np.array_split(y_train, num_cores)

results = Parallel(n_jobs=num_cores)(
    delayed(train_model)(X_chunk, y_chunk) for X_chunk, y_chunk in zip(X_train_chunks, y_train_chunks)
)


merged_history = {}
best_val_loss = float('inf')
best_val_accuracy = 0.0

for i, history in enumerate(results):
    print(f"Training Chunk {i + 1} History:")
    for key, value in history.history.items():
        merged_history_key = f"chunk_{i + 1}_{key}"
        merged_history[merged_history_key] = value
        print(f"{key}: {value}")

        if key.startswith('val_loss'):
            best_val_loss = min(best_val_loss, min(value))
        elif key.startswith('val_accuracy'):
            best_val_accuracy = max(best_val_accuracy, max(value))

print("\nBest Validation Loss:", best_val_loss)
print("Best Validation Accuracy:", best_val_accuracy)
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np
from joblib import Parallel, delayed
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf

def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model



num_runs = 10
sequential_times = []
parallel_times = []
gpu_times = []

for _ in range(num_runs):
    start_time_sequential = time.time()

    early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
    model_sequential = create_model()
    history_sequential = model_sequential.fit(X_train, y_train, epochs=50, batch_size=8,
                                              validation_data=(X_test, y_test), callbacks=[early_stop])

    end_time_sequential = time.time()
    sequential_time = end_time_sequential - start_time_sequential
    sequential_times.append(sequential_time)

    print(f"Sequential Model Training Time (Run {_ + 1}): {sequential_time:.2f} seconds")

    start_time_parallel = time.time()

    num_cores = 2
    X_train_chunks = np.array_split(X_train, num_cores)
    y_train_chunks = np.array_split(y_train, num_cores)

    results = Parallel(n_jobs=num_cores)(
        delayed(train_model)(X_chunk, y_chunk) for X_chunk, y_chunk in zip(X_train_chunks, y_train_chunks)
    )

    end_time_parallel = time.time()
    parallel_time = end_time_parallel - start_time_parallel
    parallel_times.append(parallel_time)

    print(f"Parallel Model Training Time (Run {_ + 1}): {parallel_time:.2f} seconds")

    start_time_gpu = time.time()

    model_gpu = create_model()
    with tf.device('/GPU:0'):
        history_gpu = model_gpu.fit(X_train, y_train, epochs=50, batch_size=8,
                                    validation_data=(X_test, y_test), callbacks=[early_stop])

    end_time_gpu = time.time()
    gpu_time = end_time_gpu - start_time_gpu
    gpu_times.append(gpu_time)

    print(f"GPU Model Training Time (Run {_ + 1}): {gpu_time:.2f} seconds")

labels = [f'Run {i+1}' for i in range(num_runs)]

plt.plot(labels, sequential_times, marker='o', linestyle='-', color='b', label='Sequential')
plt.plot(labels, parallel_times, marker='o', linestyle='-', color='r', label='Parallel')
plt.plot(labels, gpu_times, marker='o', linestyle='-', color='g', label='GPU')
plt.ylabel('Training Time (seconds)')
plt.title('Training Time Comparison: Sequential vs Parallel vs GPU (Multiple Runs)')
plt.legend()
plt.show()
```
